- ‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡∏≠‡∏á Correlation ‡∏ï‡πà‡∏≤‡∏á‡πÜ

- ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö Pearson, Spearman ‡πÅ‡∏•‡∏∞ Kendall

```python
# ‡∏ô‡∏≥‡πÄ‡∏Ç‡πâ‡∏≤ libraries ‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import pearsonr, spearmanr, kendalltau
import warnings
warnings.filterwarnings('ignore')

print("üîó Correlation Analysis - ‡∏™‡πà‡∏ß‡∏ô‡∏ó‡∏µ‡πà 1")
print("=" * 40)
```

## üìñ ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏Ç‡∏≠‡∏á Correlation

### 1. üìä **Pearson Correlation**
- ‡∏ß‡∏±‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå**‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏™‡πâ‡∏ô**
- ‡πÉ‡∏ä‡πâ‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏∞‡∏à‡∏≤‡∏¢‡πÅ‡∏ö‡∏ö‡∏õ‡∏Å‡∏ï‡∏¥
- ‡∏Ñ‡πà‡∏≤‡∏≠‡∏¢‡∏π‡πà‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á -1 ‡∏ñ‡∏∂‡∏á +1
- **‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ**: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡πà‡∏≠‡πÄ‡∏ô‡∏∑‡πà‡∏≠‡∏á, ‡∏Å‡∏≤‡∏£‡∏Å‡∏£‡∏∞‡∏à‡∏≤‡∏¢‡∏õ‡∏Å‡∏ï‡∏¥

### 2. üìà **Spearman Correlation**
- ‡∏ß‡∏±‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå**‡πÄ‡∏ä‡∏¥‡∏á‡∏≠‡∏±‡∏ô‡∏î‡∏±‡∏ö** (monotonic)
- ‡πÉ‡∏ä‡πâ‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏õ‡∏Å‡∏ï‡∏¥
- ‡∏ó‡∏ô‡∏ï‡πà‡∏≠ outliers ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ Pearson
- **‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ**: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÑ‡∏°‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏õ‡∏Å‡∏ï‡∏¥, ‡∏°‡∏µ outliers

### 3. üìâ **Kendall Tau**
- ‡∏ß‡∏±‡∏î‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÄ‡∏ä‡∏¥‡∏á‡∏≠‡∏±‡∏ô‡∏î‡∏±‡∏ö
- ‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏Ç‡∏ô‡∏≤‡∏î‡πÄ‡∏•‡πá‡∏Å
- ‡∏ó‡∏ô‡∏ï‡πà‡∏≠ outliers ‡∏°‡∏≤‡∏Å‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
- **‡πÄ‡∏°‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ**: ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡πâ‡∏≠‡∏¢, ‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏ß‡∏≤‡∏°‡πÅ‡∏°‡πà‡∏ô‡∏¢‡∏≥‡∏™‡∏π‡∏á

## üìä ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ correlation patterns ‡∏ï‡πà‡∏≤‡∏á‡πÜ
np.random.seed(42)
n = 500

# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 1: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏™‡πâ‡∏ô‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á
x1 = np.random.normal(0, 1, n)
y1 = 2 * x1 + np.random.normal(0, 0.5, n)

# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 2: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏™‡πâ‡∏ô‡∏≠‡πà‡∏≠‡∏ô
x2 = np.random.normal(0, 1, n)
y2 = 0.5 * x2 + np.random.normal(0, 2, n)

# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 3: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡πâ‡∏á (non-linear)
x3 = np.random.uniform(-3, 3, n)
y3 = x3**2 + np.random.normal(0, 2, n)

# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ä‡∏∏‡∏î‡∏ó‡∏µ‡πà 4: ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå
x4 = np.random.normal(0, 1, n)
y4 = np.random.normal(0, 1, n)

# ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
df_examples = pd.DataFrame({
    'linear_strong_x': x1,
    'linear_strong_y': y1,
    'linear_weak_x': x2,
    'linear_weak_y': y2,
    'nonlinear_x': x3,
    'nonlinear_y': y3,
    'random_x': x4,
    'random_y': y4
})

print("üìä ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á 4 ‡∏ä‡∏∏‡∏î:")
print(f"‡∏Ç‡∏ô‡∏≤‡∏î: {df_examples.shape}")
print(df_examples.head())
```

## üîç ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì Correlation

```python
def simple_correlation_demo(x, y, title):
    """
    Demo ‡∏Å‡∏≤‡∏£‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì correlation ‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢
    """
    print(f"\nüìä {title}")
    print("-" * 50)
    
    # ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì correlation 3 ‡πÅ‡∏ö‡∏ö
    pearson_r, pearson_p = pearsonr(x, y)
    spearman_r, spearman_p = spearmanr(x, y)
    kendall_r, kendall_p = kendalltau(x, y)
    
    print(f"Pearson r:  {pearson_r:>7.3f} (p = {pearson_p:.4f})")
    print(f"Spearman œÅ: {spearman_r:>7.3f} (p = {spearman_p:.4f})")
    print(f"Kendall œÑ:  {kendall_r:>7.3f} (p = {kendall_p:.4f})")
    
    # ‡∏ï‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°
    if abs(pearson_r) >= 0.7:
        strength = "‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á"
    elif abs(pearson_r) >= 0.3:
        strength = "‡∏õ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á"
    elif abs(pearson_r) >= 0.1:
        strength = "‡∏≠‡πà‡∏≠‡∏ô"
    else:
        strength = "‡πÑ‡∏°‡πà‡∏°‡∏µ"
    
    direction = "‡∏ö‡∏ß‡∏Å" if pearson_r > 0 else "‡∏•‡∏ö"
    print(f"‡∏Å‡∏≤‡∏£‡∏ï‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå{strength} ‡∏ó‡∏¥‡∏®‡∏ó‡∏≤‡∏á{direction}")
    
    return pearson_r, spearman_r, kendall_r

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• 4 ‡∏ä‡∏∏‡∏î
print("üîç ‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö Correlation Methods")
print("=" * 60)

results = {}
results['strong'] = simple_correlation_demo(
    df_examples['linear_strong_x'], df_examples['linear_strong_y'],
    "1. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏™‡πâ‡∏ô‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á"
)

results['weak'] = simple_correlation_demo(
    df_examples['linear_weak_x'], df_examples['linear_weak_y'],
    "2. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏™‡πâ‡∏ô‡∏≠‡πà‡∏≠‡∏ô"
)

results['nonlinear'] = simple_correlation_demo(
    df_examples['nonlinear_x'], df_examples['nonlinear_y'],
    "3. ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡πâ‡∏á (Non-linear)"
)

results['random'] = simple_correlation_demo(
    df_examples['random_x'], df_examples['random_y'],
    "4. ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå (Random)"
)
```

## üìä ‡∏Å‡∏≤‡∏£‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏£‡∏≤‡∏ü

```python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Å‡∏£‡∏≤‡∏ü‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('üîó ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á Correlation Patterns', fontsize=16, fontweight='bold')

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ó‡∏µ‡πà 1: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á
axes[0, 0].scatter(df_examples['linear_strong_x'], df_examples['linear_strong_y'], 
                   alpha=0.6, color='blue')
axes[0, 0].set_title(f'‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏Ç‡πá‡∏á‡πÅ‡∏Å‡∏£‡πà‡∏á\nPearson r = {results["strong"][0]:.3f}')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ó‡∏µ‡πà 2: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏≠‡πà‡∏≠‡∏ô
axes[0, 1].scatter(df_examples['linear_weak_x'], df_examples['linear_weak_y'], 
                   alpha=0.6, color='green')
axes[0, 1].set_title(f'‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡∏≠‡πà‡∏≠‡∏ô\nPearson r = {results["weak"][0]:.3f}')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ó‡∏µ‡πà 3: ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡πâ‡∏á
axes[1, 0].scatter(df_examples['nonlinear_x'], df_examples['nonlinear_y'], 
                   alpha=0.6, color='red')
axes[1, 0].set_title(f'‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå‡πÅ‡∏ö‡∏ö‡πÇ‡∏Ñ‡πâ‡∏á\nPearson r = {results["nonlinear"][0]:.3f}\nSpearman œÅ = {results["nonlinear"][1]:.3f}')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

# ‡∏Å‡∏£‡∏≤‡∏ü‡∏ó‡∏µ‡πà 4: ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå
axes[1, 1].scatter(df_examples['random_x'], df_examples['random_y'], 
                   alpha=0.6, color='gray')
axes[1, 1].set_title(f'‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏±‡∏°‡∏û‡∏±‡∏ô‡∏ò‡πå\nPearson r = {results["random"][0]:.3f}')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```
