````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô - ‡πÉ‡∏´‡πâ‡∏ú‡∏π‡πâ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á‡πÅ‡∏•‡∏∞‡∏Ç‡∏¢‡∏≤‡∏¢‡∏ï‡πà‡∏≠

def simple_data_pipeline():
    """
    ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á Data Pipeline ‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢
    ‡πÉ‡∏´‡πâ‡∏ú‡∏π‡πâ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏õ‡∏£‡∏±‡∏ö‡∏õ‡∏£‡∏∏‡∏á‡πÅ‡∏•‡∏∞‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå
    """
    print("üîß Simple Data Pipeline")
    print("-" * 30)
    
    # Step 1: Create sample data
    np.random.seed(42)
    n = 100
    
    data = {
        'loan_id': range(1, n+1),
        'amount': np.random.randint(5000, 50000, n),
        'rate': np.random.uniform(5, 20, n),
        'grade': np.random.choice(['A', 'B', 'C'], n),
        'income': np.random.randint(30000, 100000, n)
    }
    
    df = pd.DataFrame(data)
    print(f"‚úÖ ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {df.shape}")
    
    # Step 2: Add some problems
    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á
    df.loc[df.sample(5).index, 'income'] = np.nan
    # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ã‡πâ‡∏≥
    df = pd.concat([df, df.sample(3)], ignore_index=True)
    print(f"‚ö†Ô∏è ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏õ‡∏±‡∏ç‡∏´‡∏≤: ‡∏Ñ‡πà‡∏≤‡∏ß‡πà‡∏≤‡∏á {df.isnull().sum().sum()}, ‡∏ã‡πâ‡∏≥ {df.duplicated().sum()}")
    
    # Step 3: Clean data
    df = df.drop_duplicates()
    df['income'] = df['income'].fillna(df['income'].median())
    print(f"üßπ ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î: {df.shape}")
    
    # Step 4: Feature engineering
    df['loan_to_income'] = df['amount'] / df['income']
    df['risk_score'] = df['rate'] * df['loan_to_income']
    print(f"üè≠ ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ü‡∏µ‡πÄ‡∏à‡∏≠‡∏£‡πå: {df.shape[1]} ‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå")
    
    return df

# ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
sample_result = simple_data_pipeline()
print("\nüìä ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå:")
print(sample_result.head())
````
