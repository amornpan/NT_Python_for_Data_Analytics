- ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏™‡∏£‡∏∏‡∏õ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢‡πÜ
- ‡∏õ‡∏£‡∏∞‡∏¢‡∏∏‡∏Å‡∏ï‡πå‡πÉ‡∏ä‡πâ‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• Lending Club

## üìñ ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢

**Descriptive Analytics** ‡∏Ñ‡∏∑‡∏≠ ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏û‡∏∑‡πà‡∏≠ **‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡∏™‡∏¥‡πà‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡∏∂‡πâ‡∏ô‡πÅ‡∏•‡πâ‡∏ß**

### üéØ ‡∏à‡∏∏‡∏î‡∏õ‡∏£‡∏∞‡∏™‡∏á‡∏Ñ‡πå
- ‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô
- ‡∏™‡∏£‡∏∏‡∏õ‡πÅ‡∏ô‡∏ß‡πÇ‡∏ô‡πâ‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡∏¥‡∏î‡∏Ç‡∏∂‡πâ‡∏ô
- ‡πÄ‡∏ï‡∏£‡∏µ‡∏¢‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ç‡∏±‡πâ‡∏ô‡∏™‡∏π‡∏á

### ‚ùì ‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏ï‡∏≠‡∏ö‡πÑ‡∏î‡πâ
- ‡∏¢‡∏≠‡∏î‡∏Ç‡∏≤‡∏¢‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡πÅ‡∏•‡πâ‡∏ß‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà?
- ‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤‡∏°‡∏µ‡∏≠‡∏≤‡∏¢‡∏∏‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡πÄ‡∏ó‡πà‡∏≤‡πÑ‡∏´‡∏£‡πà?
- ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡πÑ‡∏´‡∏ô‡∏Ç‡∏≤‡∏¢‡∏î‡∏µ‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î?
- ‡πÄ‡∏ß‡πá‡∏ö‡πÑ‡∏ã‡∏ï‡πå‡∏°‡∏µ‡∏ú‡∏π‡πâ‡πÄ‡∏Ç‡πâ‡∏≤‡∏ä‡∏°‡∏Å‡∏µ‡πà‡∏Ñ‡∏ô?

## üõ†Ô∏è ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏°‡∏∑‡∏≠‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô

### üìä ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡πÄ‡∏ö‡∏∑‡πâ‡∏≠‡∏á‡∏ï‡πâ‡∏ô
1. **‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ (Mean)** - ‡∏Ñ‡πà‡∏≤‡∏Å‡∏•‡∏≤‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
2. **‡∏Ñ‡πà‡∏≤‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î/‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î (Max/Min)** - ‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
3. **‡∏ú‡∏•‡∏£‡∏ß‡∏° (Sum)** - ‡∏¢‡∏≠‡∏î‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
4. **‡∏à‡∏≥‡∏ô‡∏ß‡∏ô (Count)** - ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£
5. **‡πÄ‡∏õ‡∏≠‡∏£‡πå‡πÄ‡∏ã‡πá‡∏ô‡∏ï‡πå** - ‡∏™‡∏±‡∏î‡∏™‡πà‡∏ß‡∏ô‡∏ï‡πà‡∏≤‡∏á‡πÜ

```python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏Ç‡∏≠‡∏á Lending Club (‡∏à‡∏≥‡∏•‡∏≠‡∏á)
loan_amounts = [250000, 150000, 300000, 100000, 200000, 350000, 180000, 120000, 280000, 160000]
print("üè¶ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ 10 ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£:")
print(f"üìã ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {loan_amounts}")
print()

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô
print("üìä ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡πÄ‡∏ä‡∏¥‡∏á‡∏û‡∏£‡∏£‡∏ì‡∏ô‡∏≤ (Descriptive Analytics):")
print("="*50)

# 1. ‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢
mean_amount = sum(loan_amounts) / len(loan_amounts)
print(f"üí∞ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {mean_amount:,.0f} ‡∏ö‡∏≤‡∏ó")

# 2. ‡∏Ñ‡πà‡∏≤‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î‡πÅ‡∏•‡∏∞‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î
max_amount = max(loan_amounts)
min_amount = min(loan_amounts)
print(f"üèÜ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î: {max_amount:,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"üìâ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î: {min_amount:,.0f} ‡∏ö‡∏≤‡∏ó")

# 3. ‡∏ú‡∏•‡∏£‡∏ß‡∏°‡πÅ‡∏•‡∏∞‡∏à‡∏≥‡∏ô‡∏ß‡∏ô
total_amount = sum(loan_amounts)
count = len(loan_amounts)
print(f"üìà ‡∏¢‡∏≠‡∏î‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏£‡∏ß‡∏°: {total_amount:,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"üî¢ ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£: {count} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£")

# 4. ‡∏ä‡πà‡∏ß‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Range)
data_range = max_amount - min_amount
print(f"üìè ‡∏ä‡πà‡∏ß‡∏á‡∏Ç‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•: {data_range:,.0f} ‡∏ö‡∏≤‡∏ó")
```

## üìà ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•

‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏≠‡∏≠‡∏Å‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏ï‡πà‡∏≤‡∏á‡πÜ ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÅ‡∏•‡∏∞‡∏™‡∏±‡∏î‡∏™‡πà‡∏ß‡∏ô

```python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏ï‡∏≤‡∏°‡∏à‡∏≥‡∏ô‡∏ß‡∏ô

print("üìä ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ‡∏ï‡∏≤‡∏°‡∏à‡∏≥‡∏ô‡∏ß‡∏ô:")
print("="*40)

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡∏ä‡πà‡∏ß‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°
ranges = {
    '‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 150,000': 0,
    '150,000 - 250,000': 0,
    '250,000 - 350,000': 0,
    '‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 350,000': 0
}

# ‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
for amount in loan_amounts:
    if amount < 150000:
        ranges['‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 150,000'] += 1
    elif amount <= 250000:
        ranges['150,000 - 250,000'] += 1
    elif amount <= 350000:
        ranges['250,000 - 350,000'] += 1
    else:
        ranges['‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ 350,000'] += 1

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏•‡∏∏‡πà‡∏°
total_loans = len(loan_amounts)
for range_name, count in ranges.items():
    percentage = (count / total_loans) * 100
    bar = '‚ñà' * count  # ‡∏™‡∏£‡πâ‡∏≤‡∏á bar chart ‡∏á‡πà‡∏≤‡∏¢‡πÜ
    print(f"{range_name:>20}: {count:>2} ‡∏£‡∏≤‡∏¢‡∏Å‡∏≤‡∏£ ({percentage:>5.1f}%) {bar}")
```

## üè¶ ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏à‡∏≤‡∏Å Lending Club ‡∏à‡∏£‡∏¥‡∏á

‡∏°‡∏≤‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ Descriptive Analytics ‡∏Å‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ã‡∏±‡∏ö‡∏ã‡πâ‡∏≠‡∏ô‡∏Ç‡∏∂‡πâ‡∏ô

```python
# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤ Lending Club ‡∏ó‡∏µ‡πà‡∏™‡∏°‡∏ö‡∏π‡∏£‡∏ì‡πå‡∏Ç‡∏∂‡πâ‡∏ô
customers = {
    'loan_amount': [25000, 15000, 35000, 10000, 20000, 30000, 18000, 22000, 28000, 12000],
    'interest_rate': [10.5, 13.2, 8.9, 15.1, 11.8, 9.2, 12.5, 10.1, 9.8, 14.3],
    'annual_income': [60000, 45000, 80000, 35000, 55000, 70000, 48000, 62000, 75000, 40000],
    'loan_grade': ['B', 'C', 'A', 'D', 'B', 'A', 'C', 'B', 'A', 'D'],
    'employment_length': [5, 2, 8, 1, 3, 10, 2, 6, 12, 1]  # ‡∏õ‡∏µ‡∏Ç‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô
}

print("üè¶ ‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡∏π‡∏Å‡∏Ñ‡πâ‡∏≤ Lending Club")
print("="*50)

# 1. ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ
print("üí∞ ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ:")
avg_loan = sum(customers['loan_amount']) / len(customers['loan_amount'])
print(f"   ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg_loan:,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"   ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î: {max(customers['loan_amount']):,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"   ‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î: {min(customers['loan_amount']):,.0f} ‡∏ö‡∏≤‡∏ó")

# 2. ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏î‡∏≠‡∏Å‡πÄ‡∏ö‡∏µ‡πâ‡∏¢
print("\nüìà ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏î‡∏≠‡∏Å‡πÄ‡∏ö‡∏µ‡πâ‡∏¢:")
avg_rate = sum(customers['interest_rate']) / len(customers['interest_rate'])
print(f"   ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg_rate:.1f}%")
print(f"   ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î: {max(customers['interest_rate'])}%")
print(f"   ‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î: {min(customers['interest_rate'])}%")

# 3. ‡∏ß‡∏¥‡πÄ‡∏Ñ‡∏£‡∏≤‡∏∞‡∏´‡πå‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ
print("\nüíº ‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥‡∏£‡∏≤‡∏¢‡πÑ‡∏î‡πâ‡∏õ‡∏£‡∏∞‡∏à‡∏≥‡∏õ‡∏µ:")
avg_income = sum(customers['annual_income']) / len(customers['annual_income'])
print(f"   ‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {avg_income:,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"   ‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î: {max(customers['annual_income']):,.0f} ‡∏ö‡∏≤‡∏ó")
print(f"   ‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î: {min(customers['annual_income']):,.0f} ‡∏ö‡∏≤‡∏ó")

# 4. ‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏Å‡πÅ‡∏à‡∏á‡πÄ‡∏Å‡∏£‡∏î
print("\nüìä ‡∏Å‡∏≤‡∏£‡πÅ‡∏à‡∏Å‡πÅ‡∏à‡∏á‡πÄ‡∏Å‡∏£‡∏î‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ:")
grade_count = {}
for grade in customers['loan_grade']:
    grade_count[grade] = grade_count.get(grade, 0) + 1

total_customers = len(customers['loan_grade'])
for grade in sorted(grade_count.keys()):
    count = grade_count[grade]
    percentage = (count / total_customers) * 100
    print(f"   ‡πÄ‡∏Å‡∏£‡∏î {grade}: {count} ‡∏Ñ‡∏ô ({percentage:.1f}%)")
```
